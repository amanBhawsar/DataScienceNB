```python
dff=list(map(str,df['Location']))
dff
```

```python
Latitude = [] 
Longitude = []
ans = [] 
for i in range(len(dff)):
    k = dff[i].split("(")
    ans = k[1].split(")")
#     print(ans[0])
    a,b = ans[0].split(" ")
    Latitude.append(int(float(a)))
    Longitude.append(int(float(b)))
```

```python
Latitude
```

```python
ansi = Latitude + Longitude
dp = np.asarray(ansi)
dp
```

```python
dfs=pd.DataFrame(Latitude)
```

```python
dfs['1']=Longitude
```

```python
dfs.head()
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
#     print(kmeans)
    kmeans.fit(dfs)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
ndfs = np.asarray(dfs)
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_y = kmeans.fit_predict(ndfs)
plt.scatter(ndfs[:,0], ndfs[:,1])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.show()
```
